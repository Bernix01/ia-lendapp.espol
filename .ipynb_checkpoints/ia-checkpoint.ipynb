Importando el dataset:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

accepted_data = pd.read_csv("input/accepted.csv")
shape = accepted_data.shape
print("Datos: {}\nVariables: {}".format(shape[0],shape[1]))
```

```python
accepted_data.head()
```

Tenemos 151 variables con las que trabajar, el diccionario que explica cada variable es provisto por [el propio sitio de Lending Club](https://resources.lendingclub.com/LCDataDictionary.xlsx). Este diccionaro se clasifica en 3 hojas indicando una clasificación en los campos.


*   **LoanStates**: indicadores agregados al aprobar una aplicación de un préstamo.
*   **BrowseNotes**: indicadores visibles en la aplicación de un préstamo.
*   **RejectStats**: indicadores agregados al rechazar una aplicación de un préstamo.

Para el propósito del modelo usaremos los indicadores de **BrowseNotes**, son 121 indicadores por lo cual tendremos que aplicar algún metodo de selección.


```python
accepted_dictionary = pd.read_excel("https://resources.lendingclub.com/LCDataDictionary.xlsx")
accepted_dictionary.dropna(axis="index", inplace=True)
accepted_dictionary = accepted_dictionary.applymap(lambda x: x.strip())

browse_dictionary = pd.read_excel("https://resources.lendingclub.com/LCDataDictionary.xlsx",1)
browse_dictionary.dropna(axis="index", inplace=True)
browse_dictionary.set_index("BrowseNotesFile", inplace=True)
browse_dictionary = browse_dictionary.applymap(lambda x: x.strip())

print("Potential Browse Variables: {}\nPotential Accepted Variables: {}".format(browse_dictionary.shape[0],accepted_data.shape[1]))
```

```python
df = browse_dictionary["Description"].to_dict()
available_browse_variables = []
for col in accepted_data.columns:
    if col in df.keys():
        available_browse_variables.append(col)
        print("{:<30} {:<10}".format(col,df[col]))

print("\nTotal available browse variables: {}".format(len(available_browse_variables)))
```

Tenemos un total de 58 variables disponibles disponibles para la aplicación y en el dataset los cuales usaremos como entrada. 

```python
accepted_data[available_browse_variables].info(null_counts=True)
```

# Definiendo los datos de prueba

Ahora es necesario que diferenciemos un préstamo riesgoso de uno no riesgoso, los datos nos dan n estados provisto por el indicador `loan_status` 

```python
accepted_data.groupby("loan_status")["loan_status"].count()
```

Hay 8 estados, para considerar un préstamo no riesgoso seían los "Fully Paid" y "Current". Teniendo `1955068` préstamos no riesgosos y `305600` riesgosos. Estos son muchos préstamos no riesgosos vs los riesgosos, podemos balancear la cantidad si de los préstamos riesgosos consideramos

```python
good_loans =  accepted_data.query('loan_status == "Fully Paid" | loan_status == "Current" | loan_status == "Does not meet the credit policy. Status:Fully Paid"')
plt.figure(figsize=(13, 8))
good_loans["sub_grade"].sort_values().hist()
```

De aquí podemos ver que las calificaciones dadas por Lending Club a los préstamos nos pueden dar una sub clasificación y por lo tanto considerar otros préstamos como riesgosos.

```python
good_loans = good_loans.query('sub_grade < "C3"')
bad_loans = accepted_data.query('(loan_status != "Fully Paid" & loan_status != "Current" & loan_status != "Does not meet the credit policy. Status:Fully Paid") | sub_grade >= "C3"')
print("Good: {}\nRisky:{}".format(good_loans.shape[0],bad_loans.shape[0]))
```
